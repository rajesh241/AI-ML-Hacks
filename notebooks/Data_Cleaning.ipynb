```python
from pathlib import Path
import json
import boto3
import pandas as pd
import numpy as np
```

```python
homeDir = str(Path.home())
jsonConfigFile=f"{homeDir}/.libtech/awsConfig.json"
with open(jsonConfigFile) as config_file:
  config=json.load(config_file)
aws_access_key_id=config['aws_access_key_id']
aws_secret_access_key=config['aws_secret_access_key']
aws_region=config['aws_region']
aws_data_bucket=config['aws_data_bucket']
aws_base_url=config['aws_base_url']
```

```python
def awsInit(useEV=None):
  boto3.setup_default_session(
      aws_access_key_id=aws_access_key_id,
      aws_secret_access_key=aws_secret_access_key,
      region_name=aws_region
  )

  s3 = boto3.resource('s3', region_name=aws_region)
  bucket = s3.Bucket(aws_data_bucket)
  return s3,bucket
```

```python
def is_ascii(s):
    return all(ord(c) < 128 for c in s)
```

```python
s3,my_bucket=awsInit()
fileList=[]
for obj in my_bucket.objects.filter(Prefix="pds"):
    fileURL=f"{aws_base_url}{obj.key}"
    fileList.append(fileURL)
    
```

```python
sampleFile=fileList[0]
df=pd.read_csv(sampleFile)
df.head()
```
